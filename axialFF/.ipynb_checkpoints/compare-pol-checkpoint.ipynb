## w(x) modified

```python
def qx(x,a):
    wx = x**(1.0-x) * np.exp(-a * (1.0 - x)**2)
    dwx = x ** (1.0 - x) * np.exp(-a * (1.0 - x)**2) * ((1.0 - x) / x * (2.0 * a * x +1.0) - np.log(x))
    return ([1.0 - wx] * dwx * 2.0).flatten()

X = np.linspace(1e-4, 1, 1000)

nrows, ncols = 1, 1
fig = py.figure(figsize=(ncols*7,nrows*7*0.75))
ax = fig.add_subplot(nrows, ncols, 1)

c1=ax.plot(pol['NNPDF Q=1.087'][:,0], pol['NNPDF Q=1.087'][:,1] ,color='dimgray',ls='-.', linewidth=2)
c2=ax.fill_between(pol['NNPDF Q=1.087'][:,0],\
                   pol['NNPDF Q=1.087'][:,1] - pol['NNPDF Q=1.087'][:,2],\
                   pol['NNPDF Q=1.087'][:,1] + pol['NNPDF Q=1.087'][:,2],\
                   color='darkgray', alpha=0.85, linewidth=0)

t1=ax.plot(X, X * qx(X, -0.1) * 1.272, color='darkred', ls='-', linewidth=2)
t2=ax.plot(X, X * qx(X, -0.27) * 1.272, color='blue', ls='-', linewidth=2)
#t3=ax.plot(X, X * qx(X, 0) * 1.272, color='black', ls='-', linewidth=2)
#t3=ax.plot(X, X * qx(X, -0.5) * 1.272, color='magenta', ls='-', linewidth=2)


ax.set_xlim(1e-3,1)
ax.set_ylim(0,0.8)
ax.set_xscale('log')
ax.set_xlabel(r'$x$', size=25)
ax.set_ylabel(r'$x[\Delta u_+(x)-\Delta d_+(x)]$', size=25)
ax.tick_params(axis='both', which='both', direction='in', labelsize=25, pad=7)

L=[(c2, c1[0]), t1[0], t2[0]]
H=[
   r'NNPDF ($\mu=1.087\,\rm GeV$)',\
   r'$a=-0.1$',\
   r'$a=-0.27$',\
  ]
ax.legend(L, H, fontsize=17.5, frameon=False, labelspacing=0.3, loc="upper right", ncol=1, columnspacing=0.7)
#ax.text(0.15, 0.15, r'$\mu_0=1.087\,\rm GeV$', transform=ax.transAxes, size=18)
ax.text(0.1, 0.2, r'$w(x)=x^{1-x}e^{-a(1-x)^2}$', transform=ax.transAxes, size=22)

fig.tight_layout()
fig.savefig("results/a2.pdf")
```
