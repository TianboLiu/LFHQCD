## Veneziano amplitude

```python
def alphat(t, lam, a0):
    return a0 + t / (4.0 * lam)

def FF(tau, t, lam, a0):
    return beta(tau - 1, 1.0 - alphat(t, lam, a0)) / beta(tau - 1, 1.0 - a0)
```

```python
def FF_rho(tau, t):
    return FF(tau, t, 0.2873, 0.5049)

def FF_omega(tau, t):
    return FF(tau, t, 0.2873, 0.5223)

def FF_phi(tau, t):
    return FF(tau, t, 0.2873, 0.0269)
```

## plots

### equal amplitudes

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize = (ncols * 7, nrows * 7 * 0.75))

ax = fig.add_subplot(nrows, ncols, 1)
ax.plot(data['Q4F1p'][:,0], data['Q4F1p'][:,1],\
       linestyle = '', color = 'blue', marker = '.', markersize = 10)
ax.plot(data['Q4F1n'][:,0], data['Q4F1n'][:,1],\
       linestyle = '', color = 'red', marker = '.', markersize = 10)
t = np.linspace(0, -10, 100)
Frho = 0.5 * (FF_rho(3, t) + FF_rho(4, t))
Fomega = 0.5 * (FF_omega(3, t) + FF_omega(4, t))
ax.plot(-t, t ** 2 * 0.5 * (Fomega + Frho), linestyle = '-', linewidth = 2, color = 'blue')
ax.plot(-t, t ** 2 * 0.5 * (Fomega - Frho), linestyle = '-', linewidth = 2, color = 'red')
ax.set_xlim(0, 10)
ax.set_ylim(-0.5, 1)
ax.set_xlabel(r'$-t~\rm (GeV^2)$', size = 25)
ax.set_ylabel(r'$t^2 F_1(t)~\rm (GeV^4)$', size = 25)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 25)

fig.tight_layout()
fig.savefig('plots/f1-base.pdf')
```

### only valence

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize = (ncols * 7, nrows * 7 * 0.75))

ax = fig.add_subplot(nrows, ncols, 1)
ax.plot(data['Q4F1p'][:,0], data['Q4F1p'][:,1],\
       linestyle = '', color = 'blue', marker = '.', markersize = 10)
ax.plot(data['Q4F1n'][:,0], data['Q4F1n'][:,1],\
       linestyle = '', color = 'red', marker = '.', markersize = 10)
t = np.linspace(0, -10, 100)
Arho = 5.0 / 3.0
Aomega = 1.0 / 3.0
Frho = Arho * FF_rho(3, t) + (1.0 - Arho) * FF_rho(4, t)
Fomega = Aomega * FF_omega(3, t) + (1.0 - Aomega) * FF_omega(4, t)
ax.plot(-t, t ** 2 * 0.5 * (Fomega + Frho), linestyle = '-', linewidth = 2, color = 'blue')
ax.plot(-t, t ** 2 * 0.5 * (Fomega - Frho), linestyle = '-', linewidth = 2, color = 'red')
ax.set_xlim(0, 10)
ax.set_ylim(-0.5, 1)
ax.set_xlabel(r'$-t~\rm (GeV^2)$', size = 25)
ax.set_ylabel(r'$t^2 F_1(t)~\rm (GeV^4)$', size = 25)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 25)

fig.tight_layout()
fig.savefig('plots/f1-valence.pdf')
```

### valence + light sea

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize = (ncols * 7, nrows * 7 * 0.75))

ax = fig.add_subplot(nrows, ncols, 1)
ax.plot(data['Q4F1p'][:,0], data['Q4F1p'][:,1],\
       linestyle = '', color = 'blue', marker = '.', markersize = 10)
ax.plot(data['Q4F1n'][:,0], data['Q4F1n'][:,1],\
       linestyle = '', color = 'red', marker = '.', markersize = 10)
t = np.linspace(0, -10, 100)
Arho = 5.0 / 3.0
Aomega = 1.0 / 3.0
N5rho = -0.1
A5rho = 0.5
N5omega = 0.01
A5omega = 0.5
Frho = (1.0 - N5rho) * (Arho * FF_rho(3, t) + (1.0 - Arho) * FF_rho(4, t))\
    + N5rho * (A5rho * FF_rho(5, t) + (1.0 - A5rho) * FF_rho(6, t))
Fomega = (1.0 - N5omega) * (Aomega * FF_omega(3, t) + (1.0 - Aomega) * FF_omega(4, t))\
    + N5omega * (A5omega * FF_omega(5, t) + (1.0 - A5omega) * FF_omega(6, t))
ax.plot(-t, t ** 2 * 0.5 * (Fomega + Frho), linestyle = '-', linewidth = 2, color = 'blue')
ax.plot(-t, t ** 2 * 0.5 * (Fomega - Frho), linestyle = '-', linewidth = 2, color = 'red')
ax.set_xlim(0, 10)
ax.set_ylim(-0.5, 1)
ax.set_xlabel(r'$-t~\rm (GeV^2)$', size = 25)
ax.set_ylabel(r'$t^2 F_1(t)~\rm (GeV^4)$', size = 25)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 25)

fig.tight_layout()
fig.savefig('plots/f1-valence-sea.pdf')
```

### valence + light sea + strange sea

```python
nrows, ncols = 1, 1

fig = plt.figure(figsize = (ncols * 7, nrows * 7 * 0.75))

ax = fig.add_subplot(nrows, ncols, 1)
ax.plot(data['Q4F1p'][:,0], data['Q4F1p'][:,1],\
       linestyle = '', color = 'blue', marker = '.', markersize = 10)
ax.plot(data['Q4F1n'][:,0], data['Q4F1n'][:,1],\
       linestyle = '', color = 'red', marker = '.', markersize = 10)
t = np.linspace(0, -10, 100)
Arho = 5.0 / 3.0
Aomega = 1.0 / 3.0
N5rho = -0.1
A5rho = 0.5
N5omega = 0.01
A5omega = 0.5
N5phi = -0.015
Frho = (1.0 - N5rho) * (Arho * FF_rho(3, t) + (1.0 - Arho) * FF_rho(4, t))\
    + N5rho * (A5rho * FF_rho(5, t) + (1.0 - A5rho) * FF_rho(6, t))
Fomega = (1.0 - N5omega) * (Aomega * FF_omega(3, t) + (1.0 - Aomega) * FF_omega(4, t))\
    + N5omega * (A5omega * FF_omega(5, t) + (1.0 - A5omega) * FF_omega(6, t))
Fphi = N5phi * (FF_phi(5, t) - FF_phi(6, t))
ax.plot(-t, t ** 2 * 0.5 * (Fomega + Frho + Fphi), linestyle = '-', linewidth = 2, color = 'blue')
ax.plot(-t, t ** 2 * 0.5 * (Fomega - Frho + Fphi), linestyle = '-', linewidth = 2, color = 'red')
ax.set_xlim(0, 10)
ax.set_ylim(-0.5, 1)
ax.set_xlabel(r'$-t~\rm (GeV^2)$', size = 25)
ax.set_ylabel(r'$t^2 F_1(t)~\rm (GeV^4)$', size = 25)
ax.tick_params(axis = 'both', which = 'both', direction = 'in', labelsize = 25)

fig.tight_layout()
fig.savefig('plots/f1-valence-sea-strange.pdf')
```
