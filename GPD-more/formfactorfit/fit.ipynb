# Funcitons

```python
def G_D(Q2):
    return (1.0 + Q2/0.71)**(-2)
```

```python
def alphat(t, kappa, a0):
    return t / (4.0 * kappa**2) + a0
```

```python
class Parameters:    
    kappa = 0.533663
    a0rho = 0.492113
    a0omega = 0.5
    a0phi = 0.014219
    a0 = 0.5 - 0.1349770**2 / (4.0 * kappa**2)
    mup = 2.7928473446
    kap = mup - 1.0
    mun = -1.9130427
    kan = mun
    Mp = 0.938272081
    Mn = 0.939565413
    kau = 2.0 * kap + kan
    kad = kap + 2.0 * kan
    Mpi = 0.1349770
    normalweight = 1e6
    
    def __int__(self):
        pass
    
    def reset(self):
        self.kappa = Parameters.kappa
        self.a0 = Parameters.a0
    
    def sample(self):
        self.kappa = Parameters.kappa * np.random.normal(1.0, 0.05)
        self.a0 = 0.5 - Parameters.Mpi**2 / (4.0 * self.kappa**2)
    
    def setkappa(self, kappa):
        self.kappa = kappa
        self.a0 = 0.5 - Parameters.Mpi**2 / (4.0 * kappa**2)
        
        
    
Par = Parameters()
```

```python
class Coef:
    u3 = 2.0
    u4 = 0.0
    u5 = 0.0
    u6 = 0.0
    d3 = 1.0
    d4 = 0.0
    d5 = 0.0
    d6 = 0.0
    
    def __init__(self):
        pass
    
    def setvalue(self, div):
        udiv = np.sort(div[:3])
        self.u3 = udiv[0]
        self.u4 = udiv[1] - udiv[0]
        self.u5 = udiv[2] - udiv[1]
        self.u6 = 2.0 - udiv[2]
        ddiv = np.sort(div[3:])
        self.d3 = ddiv[0]
        self.d4 = ddiv[1] - ddiv[0]
        self.d5 = ddiv[2] - ddiv[1]
        self.d6 = 1.0 - ddiv[2]
    
    def setcoef(self, coef):
        self.u3 = coef[0]
        self.u4 = coef[1]
        self.u5 = coef[2]
        self.u6 = coef[3]
        self.d3 = coef[4]
        self.d4 = coef[5]
        self.d5 = coef[6]
        self.d6 = coef[7]
        
    def Print(self):
        print 'u3 = ', '%.4e'%self.u3
        print 'u4 = ', '%.4e'%self.u4 
        print 'u5 = ', '%.4e'%self.u5 
        print 'u6 = ', '%.4e'%self.u6
        print 'd3 = ', '%.4e'%self.d3 
        print 'd4 = ', '%.4e'%self.d4 
        print 'd5 = ', '%.4e'%self.d5 
        print 'd6 = ', '%.4e'%self.d6
        
coef = Coef()
```

```python
def FF(tau, Q2):
    return beta(tau - 1.0, 1.0 - alphat(-Q2, Par.kappa, Par.a0)) / beta(tau - 1.0, 1.0 - Par.a0)

def FFiv(tau, Q2):
    return beta(tau - 1.0, 1.0 - alphat(-Q2, Par.kappa, Par.a0rho)) / beta(tau - 1.0, 1.0 - Par.a0rho)

def FFis(tau, Q2):
    return beta(tau - 1.0, 1.0 - alphat(-Q2, Par.kappa, Par.a0omega)) / beta(tau - 1.0, 1.0 - Par.a0omega)

def FFs(tau, Q2):
    return beta(tau - 1.0, 1.0 - alphat(-Q2, Par.kappa, Par.a0phi)) / beta(tau - 1.0, 1.0 - Par.a0phi)
```

```python
def F1u(Q2):
    return coef.u3 * FF(3, Q2) + coef.u4 * FF(4, Q2) + coef.u5 * FF(5, Q2) + coef.u6 * FF(6, Q2)

def F1d(Q2):
    return coef.d3 * FF(3, Q2) + coef.d4 * FF(4, Q2) + coef.d5 * FF(5, Q2) + coef.d6 * FF(6, Q2)

def F1p(Q2):
    return 2.0/3.0 * F1u(Q2) - 1.0/3.0 * F1d(Q2)

def F1n(Q2):
    return 2.0/3.0 * F1d(Q2) - 1.0/3.0 * F1u(Q2)

def F2u(Q2):    
    return Par.kau * (np.sqrt(coef.u3 * coef.u4) * FF(4, Q2) + np.sqrt(coef.u5 * coef.u6) * FF(6, Q2)) / (np.sqrt(coef.u5 * coef.u6) + np.sqrt(coef.u3 * coef.u4))

def F2d(Q2):    
    return Par.kad * (np.sqrt(coef.d3 * coef.d4) * FF(4, Q2) + np.sqrt(coef.d5 * coef.d6) * FF(6, Q2)) / (np.sqrt(coef.d5 * coef.d6) + np.sqrt(coef.d3 * coef.d4))

def F2p(Q2):
    return 2.0/3.0 * F2u(Q2) - 1.0/3.0 * F2d(Q2)

def F2n(Q2):
    return 2.0/3.0 * F2d(Q2) - 1.0/3.0 * F2u(Q2)

def GEp(Q2):
    return F1p(Q2) - Q2 / (4.0 * Par.Mp**2) * F2p(Q2)

def GMp(Q2):
    return F1p(Q2) + F2p(Q2)

def GEn(Q2):
    return F1n(Q2) - Q2 / (4.0 * Par.Mn**2) * F2n(Q2)

def GMn(Q2):
    return F1n(Q2) + F2n(Q2)
```

```python
Q2l, Q2u = 0.0, np.inf
wp = 0.5
eth = 0.05

def Chi2(par):
    #coef.setvalue(par)
    coef.setcoef(par)
    #if np.logical_or(coef.u3 < 0, coef.d3 < 0):
    #    return 1e200
    result = 0.0
    result += Par.normalweight*(np.abs(coef.u3+coef.u4+coef.u5+coef.u6 - 2)**2 + np.abs(coef.d3+coef.d4+coef.d5+coef.d6 - 1)**2)
    Data = data['mu*G_Ep/G_Mp'][np.logical_and(Q2l < data['mu*G_Ep/G_Mp']['Q2'], data['mu*G_Ep/G_Mp']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['mu*G_Ep/G_Mp'], Data['error']
    thy = Par.mup * GEp(Q2) / GMp(Q2)
    result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Q2**wp)
    ##
    Data = data['G_Ep/G_D'][np.logical_and(Q2l < data['G_Ep/G_D']['Q2'], data['G_Ep/G_D']['Q2'] < 1.0)]
    Q2, value, error = Data['Q2'], Data['G_Ep/G_D'], Data['error']
    thy = GEp(Q2) / G_D(Q2)
    #result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Q2**wp)
    ##
    Data = data['G_Mp/mu/G_D'][np.logical_and(Q2l < data['G_Mp/mu/G_D']['Q2'], data['G_Mp/mu/G_D']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['G_Mp/mu/G_D'], Data['error']
    thy = GMp(Q2) / Par.mup / G_D(Q2)
    result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Q2**wp)
    ##
    Data = data['Q^4*F_1p'][np.logical_and(Q2l < data['Q^4*F_1p']['Q2'], data['Q^4*F_1p']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['Q^4*F_1p'], Data['error']
    thy = Q2**2 * F1p(Q2)
    result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Q2**wp)
    ##
    Data = data['Q^2*F_2p/F_1p'][np.logical_and(Q2l < data['Q^2*F_2p/F_1p']['Q2'], data['Q^2*F_2p/F_1p']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['Q^2*F_2p/F_1p'], Data['error']
    thy = Q2 * F2p(Q2) / F1p(Q2) / Par.kap
    result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Q2**wp)
    ##
    Data = data['F_2p/G_D'][np.logical_and(Q2l < data['F_2p/G_D']['Q2'], data['F_2p/G_D']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['F_2p/G_D'], Data['error']
    thy = F2p(Q2) / G_D(Q2) / Par.kap
    result += np.sum((thy - value)**2 / (error**2 + (value*eth)**2) * Data['Q2']**wp)
    ##
    Data = data['mu*G_En/G_Mn'][np.logical_and(Q2l < data['mu*G_En/G_Mn']['Q2'], data['mu*G_En/G_Mn']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['mu*G_En/G_Mn'], Data['error']
    thy = Par.mun * GEn(Q2) / GMn(Q2)
    result += np.sum((thy - value)**2 / (error**2))
    ##
    Data = data['G_En'][np.logical_and(Q2l < data['G_En']['Q2'], data['G_En']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['G_En'], Data['error']
    thy = GEn(Q2)
    result += np.sum((thy - value)**2 / (error**2))
    ##
    Data = data['G_Mn/mu/G_D'][np.logical_and(Q2l < data['G_Mn/mu/G_D']['Q2'], data['G_Mn/mu/G_D']['Q2'] < Q2u)]
    Q2, value, error = Data['Q2'], Data['G_Mn/mu/G_D'], Data['error']
    thy = GMn(Q2) / Par.mun / G_D(Q2)
    result += np.sum((thy - value)**2 / (error**2))
    return result
    
```
